```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
df
```

```python
df.head(1)
```

```python
fig=sns.FacetGrid(df)
fig.map(plt.scatter,'total_bill','tip')
plt
```

```python
fig=sns.FacetGrid(df)
fig.map(plt.bar,'total_bill','tip')
plt
```

```python
fig=sns.FacetGrid(df,col='sex')
fig.map(plt.scatter,'total_bill','tip')
plt
```

```python
fig=sns.FacetGrid(df,col='day')
fig.map(plt.scatter,'total_bill','tip')
plt
```

```python
fig=sns.FacetGrid(df,col='sex',hue='day')
fig.map(plt.scatter,'total_bill','tip')
plt
```

```python
fig=sns.FacetGrid(df,col='sex',hue='day')
fig.map(plt.scatter,'total_bill','tip')
fig.add_legend()
plt
```

```python
fig=sns.FacetGrid(df,col='sex',hue='day',palette='hot')
fig.map(plt.scatter,'total_bill','tip').add_legend()
plt
```

# Pairplot

```python
df1=sns.load_dataset('tips')
df1
```

```python
df1.head(1)
```

```python
sns.pairplot(df1)
plt
```

```python
sns.pairplot(df1,hue='sex')
plt
```

```python
sns.pairplot(df1,vars=['size','tip'],hue='smoker')
plt
```

```python
sns.pairplot(df1,vars=['size','tip'],hue='sex',hue_order=['Female','Male'])
plt
```
